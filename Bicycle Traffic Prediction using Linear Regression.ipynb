## Considering Daylight saving factor 

```python
## Function to calculate daylight hours

def hours_of_daylight(date, axis=23.44, latitude=47.61):
    """Compute the hours of daylight for the given date"""
    days = (date - pd.datetime(2000, 12, 21)).days
    m = (1. - np.tan(np.radians(latitude))
         * np.tan(np.radians(axis) * np.cos(days * 2 * np.pi / 365.25)))
    return 24. * np.degrees(np.arccos(1 - np.clip(m, 0, 2))) / 180.
```

```python
daily['Daylight_hrs'] = list(map(hours_of_daylight,daily.index))
```

## Keeping in consideration the weather for bicycle traffic

```python
weather.head()
```

```python
weather['TMIN'] /= 10
weather['TMAX'] /= 10

## Dividingb the weather by 10 to get actual value
```

```python
weather['Temp (C)'] = 0.5 * (weather['TMIN'] + weather['TMAX']) 

## Taking the average of MIN and MAX weather
```

```python
weather.head()
```

## Taking precipitation into consideration

```python
## To find if it was a dry day

weather['PRCP'] /= 254
weather['Dry day'] = (weather['PRCP'] == 0).astype(int) ## Returns 1 if it is a dry day i.e. PRCP = 0
```

```python
weather.head()
```

```python
daily = daily.join(weather[['PRCP','Temp (C)','Dry day']])

## joins the 3 columns - PRCP, Temp and Dry day to Daily 
```

```python
daily
```

```python
daily.index
```

```python
daily['annual'] = (daily.index - daily.index[0]).days /365

## Similar to linspace but calculates for the duration of 5 years i.e. 2012 to 2017
```

```python
daily.dropna(axis=0,how='any',inplace=True) ## Drops all the rows that contain Nan

column_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'Holidays',
                'Daylight_hrs', 'PRCP', 'Dry day', 'Temp (C)', 'annual']
X = daily[column_names]
y = daily['Total']  ## y is the actual total traffic
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=False)

## fit_intercept is false because = The weather for a particular day is computed based on the current day's condition and not previous day's values
```

```python
model.fit(X,y)
```

```python
daily['Predicted'] = model.predict(X)
```

```python
daily[['Total','Predicted']].plot(alpha = 0.5)
```
